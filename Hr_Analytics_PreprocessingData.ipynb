```python
df.drop(columns=['Over18', 'EmployeeCount', 'StandardHours','HourlyRate','MonthlyRate','DailyRate'], inplace=True)
```

```python
edu_map = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelors',
    4: 'Masters',
    5: 'Doctorate'
}

df['Education'] = df['Education'].map(edu_map)
```

```python
# Map codes to text labels
env_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map(env_map)
```

```python
df.head(2)
```

```python
# Map numeric satisfaction/involvement to text
level_map = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

df['JobInvolvement'] = df['JobInvolvement'].map(level_map)
df['JobSatisfaction'] = df['JobSatisfaction'].map(level_map)

```

```python
joblevel_map = {
    1: 'Entry',
    2: 'Experienced',
    3: 'Senior',
    4: 'Manager',
    5: 'Director' 
}
df['JobLevel'] = df['JobLevel'].map(joblevel_map)
```

```python
df.head(2)
```

```python
performance_map = {
    1: 'Low',
    2: 'Good',
    3: 'Excellent',
    4: 'Outstanding'
}

df['PerformanceRating'] = df['PerformanceRating'].map(performance_map)
```

```python
relationship_map = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(relationship_map)
```

```python
stock_map = {0: 'None', 1: 'Low', 2: 'Medium', 3: 'High'}
df['StockOptionLevel'] = df['StockOptionLevel'].map(stock_map)
```

```python
wlb_map = {
    1: 'Bad',
    2: 'Good',
    3: 'Better',
    4: 'Best'
}
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(wlb_map)
```

```python
df.rename(columns={"NumCompaniesWorked":"NumberCompaniesWorked"},inplace=True)
```

# EDA

```python
plt.hist(df['Age'],bins=15,width=1)
plt.title("Age Distribution Of Employees")
plt.xlabel('Age Range of Employees')
plt.ylabel('Count of Employess')
plt.grid(True)
plt.savefig("AgeRange.png")
plt.show()
```

```python
plt.hist(df['Age'], bins=15, rwidth=0.9)
plt.title("Age Distribution Of Employees")
plt.xlabel('Age')
plt.ylabel('Count of Employees')
plt.grid(True)
mean_age = df['Age'].mean()
median_age = df['Age'].median()
plt.axvline(mean_age, color='red', linestyle='--', label=f'Mean: {mean_age:.1f}')
plt.axvline(median_age, color='blue', linestyle=':', label=f'Median: {median_age:.1f}')
plt.legend()
plt.tight_layout()
plt.savefig("Age_Distribution.png")
plt.show()
```

```python
plt.hist(df['MonthlyIncome'],bins=15,rwidth=0.9)
plt.title("Salary Distribution Of Employees")
plt.xlabel('Monthly Salary')
plt.ylabel('Count of Employess')
plt.grid(True)
mean = df['MonthlyIncome'].mean()
median = df['MonthlyIncome'].median()
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.0f}')
plt.axvline(median, color='blue', linestyle=':', label=f'Median: {median:.0f}')
plt.legend()
plt.savefig("Salary_Distribution.png")
plt.show()
```

```python
plt.hist(df['DistanceFromHome'],bins=15,rwidth=0.9)
plt.title("DistanceFromHome Of Employees")
plt.xlabel('Distance')
plt.ylabel('Count of Employess')
mean_dist = df['DistanceFromHome'].mean()
plt.axvline(mean_dist, color='red',linestyle='--', label=f'Mean: {mean_dist:.1f}')
plt.legend()
plt.grid(True)
plt.legend()
plt.savefig("Distance.png")
plt.show()
```

```python
counts=df['Gender'].value_counts()
plt.pie(counts,labels=counts.index,autopct='%1.1f%%',explode=[0,0.05])
plt.legend()
plt.title("Gender Distribution of Employee")
plt.axis('equal')
plt.savefig("Gender_Distribution.png")
plt.show()
```

```python
bt_counts = df['BusinessTravel'].value_counts()


plt.bar(bt_counts.index, bt_counts.values, color='skyblue', edgecolor='black')
plt.title('Business Travel Distribution')
plt.grid()
plt.xlabel('Business Travel Type')
plt.ylabel('Number of Employees')
plt.savefig("BussinessTravel_Distribution.png")
plt.show()
```

```python
dp_counts = df['Department'].value_counts()

plt.bar(dp_counts.index, dp_counts.values, edgecolor='black')
plt.title('Employees per Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.grid()
plt.tight_layout()
plt.savefig("Employee_per_Department.png")
plt.show()
```

```python
wbl=df['WorkLifeBalance'].value_counts()
wbl

plt.bar(wbl.index,wbl.values)
plt.title('WorkLifeBalance Distribution')
plt.xlabel('WorkLifeBalance Index')
plt.ylabel('Count')
plt.grid()
plt.tight_layout()
plt.savefig("WBL_Distribution.png")
plt.show()
```

```python
bt_attr = df.groupby(['BusinessTravel', 'Attrition']).size().unstack(fill_value=0)
bt_attr.plot(kind='bar', stacked=True, color=['skyblue','salmon'], edgecolor='black')
plt.title('Attrition by Business Travel')
plt.xlabel('Business Travel')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.xticks(rotation=1)
plt.tight_layout()
plt.savefig("Attrition_BT.png")
plt.show()
```

```python
marital_statuses = sorted(df['MaritalStatus'].unique())

left_counts = [((df['MaritalStatus'] == s) & (df['Attrition'] == 'Yes')).sum() for s in marital_statuses]
stay_counts = [((df['MaritalStatus'] == s) & (df['Attrition'] == 'No')).sum() for s in marital_statuses]


plt.bar(marital_statuses, stay_counts, label='Stayed (No)')
plt.bar(marital_statuses, left_counts, bottom=stay_counts, label='Left (Yes)')
plt.title('Attrition by Marital Status (Counts)')
plt.xlabel('Marital Status')
plt.ylabel('Number of Employees')
plt.legend()
plt.tight_layout()
plt.savefig("Attrition_MaritalStatus.png")
plt.show()
```

```python
wlb_levels = ['Bad', 'Good', 'Better', 'Best']


left_counts = [((df['WorkLifeBalance'] == w) & (df['Attrition'] == 'Yes')).sum() for w in wlb_levels]
stay_counts = [((df['WorkLifeBalance'] == w) & (df['Attrition'] == 'No')).sum() for w in wlb_levels]


plt.bar(wlb_levels, stay_counts, label='Stayed (No)')
plt.bar(wlb_levels, left_counts, bottom=stay_counts, label='Left (Yes)')
plt.title('Attrition by Work-Life Balance (Counts)')
plt.xlabel('Work-Life Balance')
plt.ylabel('Number of Employees')
plt.legend()
plt.tight_layout()
plt.savefig("Attrtion_WBL.png")
plt.show()


```
